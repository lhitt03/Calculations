```python
import pandas as pd
import os
import numpy as np

# 2 spaces between name and -
df = pd.read_csv("CleanBelgium2023.csv")

bmass = 0.36
n_gas = 0.11
hard_coal = 0.39
h_fuel_oil = 0.12
hydro = 3.15
nuclear = 0.11
photo = 0 # solar
waste = 1.70
wind = 0
other = 0.69

total = (df["Biomass  - Actual Aggregated [MW]"] + df["Fossil Gas  - Actual Aggregated [MW]"] + df["Fossil Oil  - Actual Aggregated [MW]"]
       + df["Hydro Pumped Storage  - Actual Aggregated [MW]"] + df["Hydro Pumped Storage  - Actual Consumption [MW]"]+
       df["Hydro Run-of-river and poundage  - Actual Aggregated [MW]"] + df["Nuclear  - Actual Aggregated [MW]"] + df["Other  - Actual Aggregated [MW]"]+
       df["Solar  - Actual Aggregated [MW]"] + df["Waste  - Actual Aggregated [MW]"] + df["Wind Onshore  - Actual Aggregated [MW]"]+
       df["Wind Offshore  - Actual Aggregated [MW]"])

calc = (df["Biomass  - Actual Aggregated [MW]"] * bmass + df["Fossil Gas  - Actual Aggregated [MW]"] * n_gas + df["Fossil Oil  - Actual Aggregated [MW]"]*h_fuel_oil
       + df["Hydro Pumped Storage  - Actual Aggregated [MW]"]*hydro + df["Hydro Pumped Storage  - Actual Consumption [MW]"]*hydro +
       df["Hydro Run-of-river and poundage  - Actual Aggregated [MW]"] + df["Nuclear  - Actual Aggregated [MW]"] + df["Other  - Actual Aggregated [MW]"] * other +
       df["Solar  - Actual Aggregated [MW]"]* photo + df["Waste  - Actual Aggregated [MW]"] * waste + df["Wind Onshore  - Actual Aggregated [MW]"] * wind +
       df["Wind Offshore  - Actual Aggregated [MW]"] * wind)

#print(calc)

results = calc / total # gal/kWh

#print(results)

df.insert(13, "indirectWue", results)

print(df)

directory = '/Users/laurahitt'

file_name = 'Offsite_Belgium.csv'
full_path = os.path.join(directory, file_name)

# Step 8: Create the directory if it doesn't exist
os.makedirs(directory, exist_ok=True)

# Step 9: Save the resampled data to a new CSV file
df.to_csv(full_path, index=False)
```
