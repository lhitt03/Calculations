```python
import pandas as pd
import os
import numpy as np

# combining indirect and direct
# direct is from temperature
#indirect is from electricity generation
# Belgium

df = pd.read_csv("Direct_Belgium.csv")
df_res = pd.read_csv("Offsite_Belgium.csv")

total = df["directWue"] + df_res["indirectWue gal/kWh"]

df_res.insert(2, "totalWue gal/kWh", total)
df_res.insert(2, "directWue", df["directWue"])

directory = '/Users/laurahitt'

file_name = 'TotalWue_Belgium.csv'
full_path = os.path.join(directory, file_name)

# Step 8: Create the directory if it doesn't exist
os.makedirs(directory, exist_ok=True)

# Step 9: Save the resampled data to a new CSV file
df_res.to_csv(full_path, index=False)
```
