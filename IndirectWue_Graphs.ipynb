```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

# percentage change box plot
# Figure 6

# Step 1: Read the dataset
df = pd.read_csv('TotalWue_Belgium.csv')
df_res = pd.read_csv('TotalWue_Denmark.csv')
df1 = pd.read_csv('TotalWue_Finland.csv')
df2 = pd.read_csv('TotalWue_Germany.csv')
df3 = pd.read_csv('TotalWue_Ireland.csv')
df4 = pd.read_csv('TotalWue_Netherlands1.csv')
df5 = pd.read_csv('TotalWue_Netherlands2.csv')
df6 = pd.read_csv('TotalWue_Norway.csv')
df7 = pd.read_csv('TotalWue_UK.csv')

for df in [df, df_res, df1, df2, df3, df4, df5, df6, df7]:
    df['start_time'] = df['MTU'].str.split(' - ').str[0]
    df['start_time'] = df['start_time'].str.split(' \\(').str[0]

for df in [df, df_res, df1, df2, df3, df4, df5, df6, df7]:
    df['start_time'] = pd.to_datetime(df['start_time'], format='%d.%m.%Y %H:%M')


df = df.set_index('start_time')
df_res = df_res.set_index('start_time')
df1 = df1.set_index('start_time')
df2 = df2.set_index('start_time')
df3 = df3.set_index('start_time')
df4 = df4.set_index('start_time')
df5 = df5.set_index('start_time')
df6 = df6.set_index('start_time')
df7 = df7.set_index('start_time')

for df in [df, df_res, df1, df2, df3, df4, df5, df6, df7]:
    df['indirectWue L/kWh'] = df['indirectWue gal/kWh'] * 3.78541

for df in [df, df_res, df1, df2, df3, df4, df5, df6, df7]:
    df['WUE_percentage_change'] = abs(df['indirectWue L/kWh'].pct_change(periods=24) * 100)

for df in [df, df_res, df1, df2, df3, df4, df5, df6, df7]:
    df.dropna(subset=['WUE_percentage_change'], inplace=True)

all_direct_wue = [
    df['WUE_percentage_change'],
    df_res['WUE_percentage_change'],
    df1['WUE_percentage_change'],
    df2['WUE_percentage_change'],
    df3['WUE_percentage_change'],
    df4['WUE_percentage_change'],
    df5['WUE_percentage_change'],
    df6['WUE_percentage_change'],
    df7['WUE_percentage_change'] ]



all_indirect_wue_df = pd.concat(all_direct_wue, axis=1)
all_indirect_wue_df.columns = ["St. Ghislain, Belgium", "Fredericia, Denmark", "Hamina, Finland", "Hanau, Germany", "Dublin, Ireland",
                             "Eemshaven, Netherlands", "Middenmeer, Netherlands", "Skien, Norway", "Waltham Cross, UK"]
all_indirect_wue_melted = all_indirect_wue_df.melt(var_name="Locations", value_name="Percentage Change (%)")

fig, ax = plt.subplots(figsize=(15, 10))
sb.boxplot(
    x="Locations",
    y="Percentage Change (%)",
    data=all_indirect_wue_melted,
    orient="v",
    color="lightgreen", whis = [5, 85],
    showmeans=True, showfliers = False,
    boxprops={"facecolor": "lightgreen", "linewidth": 3},
    medianprops={"color": "black", "linewidth": 1.5},
    meanprops={"marker": "D", "markerfacecolor": "yellow", "markeredgecolor": "black"},
    whiskerprops={"linestyle": "-", "linewidth": 2, "color": "black"}
)

plt.title('Percentage Change of WUE Over 24-Hour Periods', fontsize=16)
plt.xlabel('Locations', fontsize=14)
plt.ylabel('Percentage Change (%)', fontsize=14)
plt.ylim(0, 100)
plt.grid(True)
plt.xticks(rotation=90, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
