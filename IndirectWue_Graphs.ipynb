```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('TotalWue_Belgium.csv')
df_res = pd.read_csv("TotalWue_Denmark.csv")
df1 = pd.read_csv("TotalWue_Finland.csv")
df2 = pd.read_csv("TotalWue_Germany.csv")
df3 = pd.read_csv("TotalWue_Ireland.csv")
df4 = pd.read_csv("TotalWue_Netherlands1.csv")
df5 = pd.read_csv("TotalWue_Netherlands2.csv")
df6 = pd.read_csv("TotalWue_Norway.csv")
df7 = pd.read_csv("TotalWue_UK.csv")


df['start_time'] = df['MTU'].str.split(' - ').str[0]
df_res["start_time"] = df_res["MTU"].str.split(" - ").str[0]
df1["start_time"] = df1["MTU"].str.split(" - ").str[0]
df2["start_time"] = df2["MTU"].str.split(" - ").str[0]
df3["start_time"] = df3["MTU"].str.split(" - ").str[0]
df4["start_time"] = df4["MTU"].str.split(" - ").str[0]
df5["start_time"] = df5["MTU"].str.split(" - ").str[0]
df6["start_time"] = df6["MTU"].str.split(" - ").str[0]
df7["start_time"] = df7["MTU"].str.split(" - ").str[0]


# Convert the extracted datetime strings to datetime objects
# Remove timezone information before conversion
df['start_time'] = df['start_time'].str.split(' \\(').str[0]
df_res["start_time"] = df_res["start_time"].str.split(' \\(').str[0]
df1["start_time"] = df1["start_time"].str.split(' \\(').str[0]
df2["start_time"] = df2["start_time"].str.split(' \\(').str[0]
df3["start_time"] = df3["start_time"].str.split(' \\(').str[0]
df4["start_time"] = df4["start_time"].str.split(' \\(').str[0]
df5["start_time"] = df5["start_time"].str.split(' \\(').str[0]
df6["start_time"] = df6["start_time"].str.split(' \\(').str[0]
df7["start_time"] = df7["start_time"].str.split(' \\(').str[0]


df['start_time'] = pd.to_datetime(df['start_time'], format='%d.%m.%Y %H:%M')
df_res["start_time"] = pd.to_datetime(df_res["start_time"], format='%d.%m.%Y %H:%M')
df1["start_time"] = pd.to_datetime(df1["start_time"], format='%d.%m.%Y %H:%M')
df2["start_time"] = pd.to_datetime(df2["start_time"], format='%d.%m.%Y %H:%M')
df3["start_time"] = pd.to_datetime(df3["start_time"], format='%d.%m.%Y %H:%M')
df4["start_time"] = pd.to_datetime(df4["start_time"], format='%d.%m.%Y %H:%M')
df5["start_time"] = pd.to_datetime(df5["start_time"], format='%d.%m.%Y %H:%M')
df6["start_time"] = pd.to_datetime(df6["start_time"], format='%d.%m.%Y %H:%M')
df7["start_time"] = pd.to_datetime(df7["start_time"], format='%d.%m.%Y %H:%M')

#  Set the 'start_time' column as the index
df.set_index('start_time', inplace=True)
df_res.set_index('start_time', inplace=True)
df1.set_index('start_time', inplace=True)
df2.set_index('start_time', inplace=True)
df3.set_index('start_time', inplace=True)
df4.set_index('start_time', inplace=True)
df5.set_index('start_time', inplace=True)
df6.set_index('start_time', inplace=True)
df7.set_index('start_time', inplace=True)


# Convert units
df["indirectWue L/kWh"] = df["indirectWue gal/kWh"] * 3.78541
df_res["indirectWue L/kWh"] = df_res["indirectWue gal/kWh"] * 3.78541
df1["indirectWue L/kWh"] = df1["indirectWue gal/kWh"] * 3.78541
df2["indirectWue L/kWh"] = df2["indirectWue gal/kWh"] * 3.78541
df3["indirectWue L/kWh"] = df3["indirectWue gal/kWh"] * 3.78541
df4["indirectWue L/kWh"] = df4["indirectWue gal/kWh"] * 3.78541
df5["indirectWue L/kWh"] = df5["indirectWue gal/kWh"] * 3.78541
df6["indirectWue L/kWh"] = df6["indirectWue gal/kWh"] * 3.78541
df7["indirectWue L/kWh"] = df7["indirectWue gal/kWh"] * 3.78541

#  Calculate percentage change over 24-hour periods
df['WUE_percentage_change'] = abs(df['indirectWue L/kWh'].pct_change(periods=24) * 100)
df_res["WUE_percentage_change"] = abs(df_res["indirectWue L/kWh"].pct_change(periods=24) * 100)
df1["WUE_percentage_change"] = abs(df1["indirectWue L/kWh"].pct_change(periods=24) * 100)
df2["WUE_percentage_change"] = abs(df2["indirectWue L/kWh"].pct_change(periods=24) * 100)
df3["WUE_percentage_change"] = abs(df3["indirectWue L/kWh"].pct_change(periods=24) * 100)
df4["WUE_percentage_change"] = abs(df4["indirectWue L/kWh"].pct_change(periods=24) * 100)
df5["WUE_percentage_change"] = abs(df5["indirectWue L/kWh"].pct_change(periods=24) * 100)
df6["WUE_percentage_change"] = abs(df6["indirectWue L/kWh"].pct_change(periods=24) * 100)
df7["WUE_percentage_change"] = abs(df7["indirectWue L/kWh"].pct_change(periods=24) * 100)

# Drop NaN values
df.dropna(subset=['WUE_percentage_change'], inplace=True)
df_res.dropna(subset=['WUE_percentage_change'], inplace=True)
df1.dropna(subset=['WUE_percentage_change'], inplace=True)
df2.dropna(subset=['WUE_percentage_change'], inplace=True)
df3.dropna(subset=['WUE_percentage_change'], inplace=True)
df4.dropna(subset=['WUE_percentage_change'], inplace=True)
df5.dropna(subset=['WUE_percentage_change'], inplace=True)
df6.dropna(subset=['WUE_percentage_change'], inplace=True)
df7.dropna(subset=['WUE_percentage_change'], inplace=True)

fig, ax = plt.subplots(figsize=(15, 10))

all_indirect_wue = [df["WUE_percentage_change"], df_res["WUE_percentage_change"], df1["WUE_percentage_change"], df2["WUE_percentage_change"], df3["WUE_percentage_change"],
                  df4["WUE_percentage_change"], df5["WUE_percentage_change"], df6["WUE_percentage_change"], df7["WUE_percentage_change"]]


medianprops = dict(color = "black", linewidth = 1.5)
boxprops = dict(facecolor = "lightgreen", linewidth = 3)
whiskerprops = dict(linestyle = '-', linewidth = 3)
meanprops = {"marker": "D", "markerfacecolor":"yellow", "markeredgecolor":"black" }

ax.boxplot(all_indirect_wue, vert = True, patch_artist = True, showmeans = True, showfliers = False, boxprops = boxprops, medianprops = medianprops, meanprops = meanprops, whiskerprops = whiskerprops )

labels = ["St. Ghislain, BE", 'Fredericia, DK', 'Hamina, FI', 'Hanau, DE (Germany)', 'Dublin, IE', 'Eemshaven, NE', 'Middenmeer, NE', 'Skien, NO', 'Waltham Cross, UK']

plt.xticks(range(1, len(all_indirect_wue) + 1), labels)
plt.xticks(rotation = 90, ha = 'right', fontsize = 14)
plt.yticks(fontsize = 14)


plt.title('Percentage Change of WUE Over 24-Hour Periods', fontsize = 16)
plt.ylabel('Percentage Change (%)', fontsize = 14)
plt.xlabel('Locations', fontsize = 14)
ax.set_ylim(0, 140)  # Adjust y-axis limits to ensure all data is visible
plt.grid(True)

plt.show()
```
